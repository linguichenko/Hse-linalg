```python
###ВАШ КОД ЗДЕСЬ###
filename ='/content/drive/MyDrive/textsSirius/корпусвесь.txt'
with open (filename, 'r', encoding='utf-8', errors = 'ignore') as file:
    text = file.readlines()
list_words = []
cnt = 0
ans = 0
for tmp in text:
    words = tmp.split()
    ans += len(words)
    for i in range (len(words)):
        if words[i] == 'время':
          cnt += 1
          if i == 0:
            list_words.append(words[1])
          elif i == len(words) - 1:
            list_words.append(words[len(words) - 2])
          else:
            list_words.append(words[i - 1])
            list_words.append(words[i + 1])

print(*list_words)
print(cnt)
print(ans)








###на входе - список слов, стоящих рядом со словом таргета###
```

> В питоне есть такой модуль - специально для того, чтобы считать сколько раз элемент встретился в списке - Counter, он находится в модуле collections. Ниже пример использования, на входе подаете список, на выходе - получаете квази частотный словарь.

```python
# подсчитаем, сколько раз встретились слова из списка в нем

#импортим модуль
from collections import Counter
сount_words = Counter(list_words)
print(сount_words) #все совпадает по частотности
```

Не в том, в котором нужно 😟😟😟 Они никак не упорядочены по частотности... Чтобы получить слова в порядке возрастания, нужно воспользоваться функцией most_common. Словарь превратится в список пар (слово, частотность), причем упорядоченных от большей частотности к меньшей.


```python
listik = []
for element in сount_words.most_common(507):
  listik.append(list(element))
print(listik)


  # к получившимся элементам можно обращаться по индексу!
```
